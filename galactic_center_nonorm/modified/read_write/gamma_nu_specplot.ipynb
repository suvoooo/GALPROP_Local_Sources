### Check $\gamma$ and $\nu$ Specrta 
Proton propagate near galactic center and interacts with the gas distribution near molecular clouds, resulting 
in gamma and neutrino production. 
The two most important processes are -- 
* $\pi ^0 \rightarrow \gamma \gamma $. 
* $\pi \rightarrow \mu \nu _{\mu}$. This leads to $\mu \rightarrow e\nu _{\mu} \nu _{e}$. 

Here we compare the $\gamma$ and $e_{\nu}, \mu _{\nu}$ spectra produced by High Energy protons from a past hypernova interacting with the central molecular clouds..  

```python
import math 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
### Gamma section 

def readgammaTaue4():
    gTe4file = open('./gamma_spec_check3DXYZ2EhighTau1e4.d', 'r')
    headerline1 = gTe4file.readline()
    headerline2 = gTe4file.readline()
    gTe4vals = {}
    while True:
        stringline = gTe4file.readline()
        if stringline=='':
            break
        stringlist = stringline.split()
        E = float(stringlist.pop(0))/(10**3)
        flux = float(stringlist.pop(0))
#         totflux = float(stringlist.pop(0))
        gTe4vals[E] = [flux]
    return gTe4vals





def readgammaTaue5():
    gTe5file = open('./gamma_spec_check3DXYZ2EhighTau1e5.d', 'r')
    headerline1 = gTe5file.readline()
    headerline2 = gTe5file.readline()
    gTe5vals = {}
    while True:
        stringline = gTe5file.readline()
        if stringline=='':
            break
        stringlist = stringline.split()
        E = float(stringlist.pop(0))/(10**3)
        flux = float(stringlist.pop(0))
#         totflux = float(stringlist.pop(0))
        gTe5vals[E] = [flux]
    return gTe5vals



def plotgammaTaue4(fig, gTe4vals=readgammaTaue4()):
    E = []
    piflux = []
    totflux = []
    for point in sorted(gTe4vals.keys()):
        E.append(point)
        piflux.append(gTe4vals[point][0] * (point**2))
#         totflux.append(gTe4vals[point][1] * (point**2))
    plt.plot(E, piflux, linestyle='-', color='orange', alpha=0.7, label=r'$\tau: 1e4, \gamma$ Flux')

    
    
def plotgammaTaue5(fig, gTe5vals=readgammaTaue5()):
    E = []
    piflux = []
    totflux = []
    for point in sorted(gTe5vals.keys()):
        E.append(point)
        piflux.append(gTe5vals[point][0] * (point**2))
#         totflux.append(gTe5vals[point][1] * (point**2))
    plt.plot(E, piflux, linestyle='-.', color='magenta', alpha=0.7, label=r'$\tau: 1e5, \gamma$ Flux')
```
